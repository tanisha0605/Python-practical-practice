```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime

# Function to read data from CSV file
def read_data(filename):
    dates = []
    opens = []
    highs = []
    lows = []
    closes = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            date = datetime.strptime(row['Date'], '%m-%d-%y')
            open_price = float(row['Open'])
            high_price = float(row['High'])
            low_price = float(row['Low'])
            close_price = float(row['Close'])
            dates.append(date)
            opens.append(open_price)
            highs.append(high_price)
            lows.append(low_price)
            closes.append(close_price)
    return dates, opens, highs, lows, closes

# Main function to plot line chart
def plot_line_chart(dates, opens, highs, lows, closes):
    plt.plot(dates, opens, linestyle='-', label='Open')
    plt.plot(dates, highs, linestyle='-', label='High')
    plt.plot(dates, lows, linestyle='-', label='Low')
    plt.plot(dates, closes, linestyle='-', label='Close')
    plt.title('Alphabet Inc. - Financial Data (Oct 3, 2016 - Oct 7, 2016)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# Read data from CSV
dates, opens, highs, lows, closes = read_data('fdata.csv')

# Plot line chart
plot_line_chart(dates, opens, highs, lows, closes)

```
