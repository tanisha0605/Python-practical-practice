```python
total_calls=(
    df["Total day calls"]+df["Total eve calls"]+df["Total night calls"]+df["Total intl calls"]
)
df.insert(loc=len(df.columns),column="Total calls",value=total_calls)
df.head()
```

```python
df["Total charge"]=(
    df["Total day calls"]+df["Total eve calls"]+df["Total night calls"]+df["Total intl calls"]
)
```

```python
df.drop(["Total charge","Total calls"],axis=1,inplace=False)
```

```python
pd.crosstab(df["Churn"],df["International plan"],margins=True)
```
