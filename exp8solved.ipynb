BAR GRAPHS

```python
Divisions=["Div-A","Div-B","Div-C","Div-D","Div-E"]
Divisions_Average_Marks=[70,82,48,60,98]
plt.figure(figsize=(5,5))
plt.bar(Divisions,Divisions_Average_Marks,color="red")
plt.title("Average Marks for each Division")
plt.xlabel("Divisions")
plt.ylabel("Average Marks")
plt.show()
```

```python
Divisions=["Div-A","Div-B","Div-C","Div-D","Div-E"]
Divisions_Average_Marks=[70,82,48,60,98]
variance=[5,8,7,2,4]
plt.figure(figsize=(5,5))
plt.barh(Divisions,Divisions_Average_Marks,xerr=variance,color="red")
plt.title("Average Marks for each Division")
plt.xlabel("Divisions")
plt.ylabel("Average Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Divisions=["Div-A","Div-B","Div-C","Div-D","Div-E"]
Divisions_Average_Marks=[70,82,48,60,98]
Boys_Average_Marks=[64,80,45,50,90]

index=np.arange(5)
width=0.30


plt.bar(index,Divisions_Average_Marks,width,color="green",label="Division Average Marks")
plt.bar(index+width,Boys_Average_Marks,width,color="blue",label="Boys Average Marks")
plt.title("Marks")

plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index+width/2,Divisions)
plt.legend(loc="best")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Divisions=["Div-A","Div-B","Div-C","Div-D","Div-E"]
Girls_Average_Marks=[70,82,48,60,98]
Boys_Average_Marks=[64,80,45,50,90]

index=np.arange(5)
width=0.30


plt.bar(index,Girls_Average_Marks,width,color="pink",label="Girls Average Marks",bottom=Boys_Average_Marks)
plt.bar(index,Boys_Average_Marks,width,color="blue",label="Boys Average Marks")
plt.title("Marks")

plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.xticks(index,Divisions)
plt.legend(loc="best")
plt.show()
```

PIE CHARTS

```python
firms=["Firm A","Firm B","Firm C","Firm D","Firm E"]
market_share=[20,25,15,10,20]
explode=[0,0.1,0,0,0]

plt.pie(market_share,labels=firms,explode=explode,shadow=True,startangle=45)
plt.title("Market Share of Top 5 Firms")
plt.axis('equal')
plt.legend(title='List of Firms',loc='best')
plt.show()
```

HISTOGRAM
